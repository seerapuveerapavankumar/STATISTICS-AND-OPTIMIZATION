## *Question*

A training manager wondered whether the length of time his trainees revised for an examination had any effect on the marks they scored in the examination.
Before the exam, he asked a random sample of them to honestly estimate how long, to the nearest hour, they had spent revising.  After the examination he investigated the relationship between the two variables.


Trainee	A	B	C	D	E	F	G	H	I	J

Revision time	4	9	10	14	4	7	12	22	1	17

Exam mark	31	58	65	73	37	44	60	91	21	84

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings('ignore')
```

```python
# Dataset

data = {'Trainee' :	['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],

'Revision_Time' : [4, 9, 10, 14, 4, 7, 12, 22, 1, 17],

'Exam_Mark'	: [31, 58, 65, 73, 37, 44, 60, 91, 21, 84]}
```

```python
# Create a DataFrame

data = pd.DataFrame(data)

X = data[['Revision_Time']]  # Independent variable
Y = data['Exam_Mark']        # Dependent variable
```

### a)	Plot the scatter diagram in order to inspect the data.

```python
plt.scatter(X, Y, color='blue', label='Data Points')
plt.title('Scatter Plot of Revision Time vs Exam Mark')
plt.xlabel('Revision Time (Hours)')
plt.ylabel('Exam Mark')
plt.legend()
plt.show()
```

### b)	Calculate the correlation coefficient.

```python
correlation_coefficient = data['Revision_Time'].corr(data['Exam_Mark'])

print(f"Correlation Coefficient (r): {correlation_coefficient:.2f}")
```

### c)	Calculate the regression model to fit the data.

```python
model = LinearRegression()

model.fit(X, Y)
slope = model.coef_[0]
intercept = model.intercept_
print(f"Regression Equation: y = {slope:.2f}x + {intercept:.2f}")
```

### d)	Plot the regression line on the scatter diagram.

```python
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.title('Regression Line on Scatter Plot')
plt.xlabel('Revision Time (Hours)')
plt.ylabel('Exam Mark')
plt.legend()
plt.show()
```
